```python
img = cv2.imread("1.jpg")
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
import matplotlib.pyplot as plt
# Display the image using Matplotlib
plt.imshow(img_rgb)
plt.axis('off')  # Turn off axis labels
plt.show()
```

```python
img2 = cv2.imread("2.jpg")
```

```python
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
plt.axis('off')  # Turn off axis labels
plt.show()
```

```python
##Image convert from RGB to Grayscale

rgb_to_gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
plt.imshow(rgb_to_gray)
plt.axis('off')  # Turn off axis labels
plt.show()
```

```python
##Image convert from Grayscale to RGB  

gray_to_rgb = cv2.cvtColor(rgb_to_gray, cv2.COLOR_GRAY2RGB)
plt.imshow(gray_to_rgb)
plt.axis('off')  # Turn off axis labels
plt.show()
```
