```python
import cv2
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# image load
img = cv2.imread('1.jpg')
```

```python
# Display img


plt.imshow(img)
plt.axis('off')  # Turn off axis labels
plt.show()
```

```python
#origional image 

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis('off')  # Turn off axis labels
plt.show()
```

```python
# rgb to gray

rgbtogray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
plt.imshow(rgbtogray)
plt.axis('off')  # Turn off axis labels
plt.show()
```
