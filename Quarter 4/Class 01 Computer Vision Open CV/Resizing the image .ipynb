```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("2.jpg")
plt.imshow(img)
plt.axis("off")
plt.show
```

```python
img.shape
```

```python
# resize from origional to 128 px

resize_image = cv2.resize(img, (128, 128))
```

```python
plt.imshow(resize_image)
plt.axis("off")
plt.show
```

```python
cv2.imwrite("Size128.jpg", resize_image)
```

```python
# resize from 128 px to 800px
resize_image_InHigherPixels = cv2.resize(resize_image, (3672, 4896))
plt.imshow(resize_image_InHigherPixels)
plt.axis("off")
plt.show
```

```python
cv2.imwrite("Size3565.jpg", resize_image_InHigherPixels)
```
